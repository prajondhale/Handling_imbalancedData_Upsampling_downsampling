Upsampling 

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1    # class_0 is 900 , class_1 is 100 , 

####  upsampling using this try to make class_1 900 and class_0 100 so both get balanced
```

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop = True)
print(df)    
```

```python
df['target'].value_counts()
```

```python
## apply upsampling 

df_minority = df[df['target']==1]
df_majority = df[df['target'] == 0]
```

```python
### use library  : resample to create more sample for minory one

from sklearn.utils import resample

df_minority_unsampled = resample(df_minoriy, replace = True,
                                   n_samples = len(df_majority), random_state = 42)
```

```python
df_minority_unsampled.shape
```

```python
df_minority_unsampled.head()
```

```python

df_upsampled = pd.concat([df_majority, df_minority_unsampled])
```

```python
df_upsampled['target'].value_counts() 
 ## here now 0 = 900, 1= 900 so minority also changes to 900 

##  so here both get balanced by upsampling
```

### Downsampling

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

```

```python
n_class_0,n_class_1 
```

```python
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop = True)
print(df) 
```

```python
print(df['target'].value_counts())
```

```python
## apply Downsampling : 

df_minority = df[df['target']==1]
df_majority = df[df['target'] == 0]
```

```python
### use library  : resample to reduce the sample from majority by replace = False

from sklearn.utils import resample

df_majority_downsampled = resample(df_majority, replace = False,
n_samples = len(df_majority), random_state = 42)
```

```python
df_downsampled = pd.concat([df_majority, df_majority_downsampled])
```
